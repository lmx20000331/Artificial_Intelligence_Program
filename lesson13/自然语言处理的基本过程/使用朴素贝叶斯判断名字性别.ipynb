````python
path = './dataset/'
file_path = path + 'charfreq.csv'
````

````python
male_total = 0
female_total = 0
freq = {}    #字典
````

## 字典参照：```https://www.runoob.com/python/python-dictionary.html```
## 元组参照：```https://www.runoob.com/python/python-tuples.html```

````python
# 一共30分
with open(file_path, 'r', encoding='utf8' ) as f:  # 打开文件（只读，utf-8）  5分  
    first_line = True
    freq = {}
    for line in f:
        #print(line)
        if first_line is True:
            first_line = False
            continue
        char, male, female = line.split(',')    # 读取每一行，使用','进行分割 5分
        male_total +=  int(male)           # 统计male列的总数（需要进行类型转换）5分
        female_total += int(female)          # 统计female列的总数（需要进行类型转换）5分
        
        # 此处添加一行代码：freq字典以char为键,值是由female和male组成的元组    10分
        male_female_total_dict = (int(male), int(female))
        freq[char] = male_female_total_dict
````

````python
freq
````

````python
total = male_total + female_total
total
````

````python
# 一共15分
for char in freq: # 遍历freq
    male, female = freq[char]
    male = float(male/male_total)
    female = float(female/female_total)
    #分析female在female_total中的占比，male在male_total中的占比，最后依旧以数组的形式写回值中（参照字典链接）。15分
    freq[char] = (male, female)  
````

````python
def prob_for_gender(firstname, gender=1 ):  #  定义prob_for_gender函数，参数为：firstname，gender=0（默认性别为女）5分
    
    if gender == 1:      # 计算名字P(gender=女)的概率    5分
        p = 1. *(female_total/total)
    else:                # 计算名字P(gender=男)的概率
        p = 1. *(male_total/total)
        
    # 使用贝叶斯计算概率P         10分
    # 举个例子：P(name=本山|gender=男) = P(name has 本|gender=男) * P(name has 山|gender=男)
    for char in firstname:        
        p *= freq.get(char, (0, 0))[gender] 

    return p
````

````python
def bys_guess(name):
    first_name = name[1:]   #定义除了姓之外的其它字为first_name
    
    #此处写两行代码：判断first_name中的每个字是否为中文（提示：中文在u'\u4e00' <= x<= u'\u9fa0'之间）     5分
    for char in first_name:
        assert u'\u4e00' <= char <= u'\u9fa0' # u'姓名必须为中文'
    
    pf = prob_for_gender(first_name, 1)     #得出属于female的概率pf          5分
    pm = prob_for_gender(first_name, 0)     #得出属于male的概率pm            5分
    
    
    # 此处添加若干行代码  
    # 1. 如果属于male的概率大，返回（'male',  1. * pm / (pm + pf))  如果属于female的概率大：('female', 1. * pf / (pm + pf))      5分
    if pf > pm:
        return ('female', 1. *pf /(pm+pf))
    elif pf < pm:
        return ('male', 1. *pm /(pm+pf))
    else:
        return ('unknown', 0)
    # 2. 其它情况返回 ('unknown', 0)   5分
````

````python
# 此处添加一行代码： 调用 bys_guess函数 猜测“张丹丹”是男是女的概率  5分
print('张丹丹：', bys_guess('张丹丹'))
````
